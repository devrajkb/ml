# Import module and crate data genearator

```python
import numpy
import random
import scipy
from scipy.spatial import distance
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib
import matplotlib.pyplot as plt
import sklearn
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering


def get_dataset_using_sklearn_make_blobs(n_samples=20, centers=2):
    X, y = make_blobs(n_samples = n_samples, n_features = 2, centers = centers , cluster_std = 1, random_state=0)
    return X,y


```

# Visualize dendogram with different number of clusters 

```python
for k in range(2,5):
    data, y = get_dataset_using_sklearn_make_blobs(1000, k)
    print("dataset samples, features =", data.shape, "number of cluster =", k)
    #print(data)
    print("\ndata plot: \n")
    plt.scatter(x=data[:,0], y=data[:,1], c=y, cmap='rainbow')
    plt.show()

    X=data[:,0:2]
    
    linked = linkage(X, method='ward', metric='euclidean')
    print("\ndendrogram plot: \n")
    dendrogram(linked,
                orientation='top',
                labels=y,
                distance_sort='descending')
    plt.show()
    clustering = AgglomerativeClustering(n_clusters = k, affinity='euclidean', linkage='ward')
    clustering.fit_predict(X)
    print("clusters ", clustering.n_clusters_)
    #print("childerns ",clustering.children_.shape)
    #print("Distance ",clustering.distances_)
    #print("distance ",clustering.labels_)
    print("\nPredicated cluster plot: \n")
    plt.scatter(data[:,0], data[:,1], c=clustering.labels_, cmap='rainbow')
    plt.show()
```
